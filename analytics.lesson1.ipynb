## задание 2

```python
keys = ['One', 'Two', 'Three']
values = [1, 2, 3]
```

### 2.1 for

```python
def get_dict(keys: list, values: list) -> dict:
    dict3 = {}
    for index in range(len(keys)):
        dict3[keys[index]] = values[index]
    return dict3

print(get_dict(keys, values))
```

### 2.2 comprehensions

```python
def get_dict_c(keys: list, values: list) -> dict:
    return {keys[i]: values[i] for i in range(len(keys))}
    
print(get_dict_c(keys, values))
```

## задание 3

## 3.1 for

```python
def get_by_keys(dict1: dict, keys: list = []) -> dict:
    dict2 = {}
    for key in keys:
        dict2[key] = dict1[key]
    return dict2

get_by_keys(dict1, ['One', 'Two'])
```

## 3.2 c

```python
def get_by_keys_c(dict1: dict, keys: list = []) -> dict:
    return {key: dict1[key] for key in keys}

get_by_keys(dict1, ['One', 'Two'])
```

### задача 4

```python
from datetime import datetime

def calc_income_by_month(dates: list, incomes: list) -> dict:
    d = {}
    for index in range(len(dates)):
        month = str(datetime.strptime(dates[index], "%Y-%m-%d").date().month)
        income = incomes[index]
        d[month] = d.get(month, 0) + income
    return d

calc_income_by_month(dates, income)
```
