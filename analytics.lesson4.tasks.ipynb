## 3

```python
plt.figure(figsize=(10,6))

sns.boxplot(data=df, y='price', x = 'grade', whis=1.5)
plt.title('Взаимосвязь стоимости недвижимости и класса жилья')
plt.ylabel('price')
plt.xlabel('grade')
```

На стоимость напрямую влияет класс жилья

```python
plt.figure(figsize=(10,6))

sns.boxplot(data=df, y='price', x = 'bathrooms', whis=1.5)
plt.title('Взаимосвязь стоимости недвижимости и кол-ва ванных комнат')
plt.ylabel('price')
plt.xlabel('bathrooms')

plt.xticks(rotation=50)
''
```

Наблюдается связь кол-ва ваннхы комнат с ценой, правда в районе 6.25-7.5 ванных комнат есть поадение цены

```python
plt.figure(figsize=(10,6))

sns.boxplot(data=df, y='price', x = 'floors', whis=1.5)
plt.title('Взаимосвязь стоимости недвижимости и кол-вом этажей')
plt.ylabel('price')
plt.xlabel('floors')
```

видно, что максиальная цена получается при высоте в 2.5 этажа

```python
plt.figure(figsize=(10,6))

sns.boxplot(data=df, y='price', x = 'waterfront', whis=1.5)
plt.title('Взаимосвязь стоимости недвижимости и наличием вида на воду')
plt.ylabel('price')
plt.xlabel('waterfront')
```

Цена напрямую зависит от наличия водоема, не считая единичных выбросов

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(6,4))

sns.heatmap(corr_matrix, annot=True, linewidths = 0.5, cmap='coolwarm')
```

Здесь видно, что на цену влияют такие параметры, как: bedrooms, floors, waterfront, sqft_basement, lat, view, bathrooms, qrft_above, sqft_living15, sqft_living, grade
